```python
import cobra
```

```python
from cobra import Model,Metabolite,Reaction
```

```python
model =Model("Model3")
```

```python
#########  Glu+ATP========> G6P + ADP+H
HK=Reaction("HK")
HK.name="Hexokinase"
HK.lower_bound=0
HK.upper_bound=1000

######### G6P<===========> F6P
PGI=Reaction("PGI")
PGI.name="Glucose 6-phosphate isomerase"
PGI.lower_bound=-1000
PGI.upper_bound=1000

####  F6P + ATP → FDP + ADP + H
PFK=Reaction("PFK")
PFK.name="Phosphofructokinase"
PFK.lower_bound=0
PFK.upper_bound=1000

####### DHAP <====> GAP
TPI=Reaction("TPI")
TPI.name="Triose-phosphate isomerase"
TPI.lower_bound=-1000
TPI.upper_bound=1000

#####  FDP <====> DHAP + GAP
ALD=Reaction("ALD")
ALD.name="Fructose 1,6-diphosphate aldolase"
ALD.lower_bound=-1000
ALD.upper_bound=1000

##### GAP + NAD + Pi <=====>  DPG13 + NADH + H
GAPDH=Reaction("GAPDH")
GAPDH.name="Glyceraldehyde 3-phosphate dehydrogenase"
GAPDH.lower_bound=-1000
GAPDH.upper_bound=1000

##### DPG13 + ADP  <=====>  PG3 + ATP
PGK=Reaction("PGK")
PGK.name="Phosphoglycerate kinase"
PGK.lower_bound=-1000
PGK.upper_bound=1000

#####  PG3 <====> PG2
PGLM=Reaction("PGLM")
PGLM.name="Phosphoglycerate mutase"
PGLM.lower_bound=-1000
PGLM.upper_bound=1000

######  PG2 <=====> PEP + H2O
ENO=Reaction("ENO")
ENO.name="Enolase"
ENO.lower_bound=-1000
ENO.upper_bound=1000

#####  PEP + ADP + H+ → PYR + ATP
PK=Reaction("PK")
PK.name="Pyruvate kinase"
PK.lower_bound=0
PK.upper_bound=1000

##### PYR + NADH + H+ <====> LAC + NAD
LDH=Reaction("LDH")
LDH.name="Lactate dehydrogenase"
LDH.lower_bound=-1000
LDH.upper_bound=1000

#####  AMP →
AMP=Reaction("AMP")
AMP.name="AMP export"
AMP.lower_bound=0
AMP.upper_bound=1000

#### 2ADP <====> AMP + ATP
APK=Reaction("APK")
APK.name="Adenylate kinase"
APK.lower_bound=-1000
APK.upper_bound=1000

##### PYR <=====>
PYR=Reaction("PYR")
PYR.name="Pyruvate exchange"
PYR.lower_bound=-1000
PYR.upper_bound=1000

#####  LAC <=====>
LAC=Reaction("LAC")
LAC.name="Lactate exchange"
LAC.lower_bound=-1000
LAC.upper_bound=1000

#### ATP + H2O → ADP + Pi + H
ATP=Reaction("ATP")
ATP.name="ATP hydrolysis"
ATP.lower_bound=0
ATP.upper_bound=1000


######  NADH → NAD + H
NADH=Reaction("NADH")
NADH.name="NADH oxidation"
NADH.lower_bound=0
NADH.upper_bound=1000


# Gluc in the cell
Gluin=Reaction("GLUin")
Gluin.name="Glucose import"
Gluin.lower_bound=1
Gluin.upper_bound=1

#### → AMP
AMPin=Reaction("AMPin")
AMPin.name="AMP import"
AMPin.lower_bound=1
AMPin.upper_bound=1

####  H <====>
H=Reaction("H")
H.name="Proton exchange"
H.lower_bound=-1000
H.upper_bound=1000

##### H2O <===>
H2O=Reaction("H2O")
H2O.name="Water exchange"
H2O.lower_bound=-1000
H2O.upper_bound=1000
```

```python
Gluc=Metabolite(
    "Gluc",
    formula="C6H12O6",
    name="Glucose",
    compartment='c')

G6P=Metabolite(
    "G6P",
    formula="C6H13O9P",
    name="Glucose 6-phosphate",
    compartment='c')

F6P=Metabolite(
    "F6P",
    formula="C6H13O9P",
    name="Fructose 6-phosphate",
    compartment='c')

FDP=Metabolite(
    "FDP",
    formula="",
    name="Fructose 1,6-diphosphate",
    compartment='c')

DHAP=Metabolite(
    "DHAP",
    formula="",
    name="Dihydroxyacetone phosphate",
    compartment='c')

GAP=Metabolite(
    "GAP",
    formula="",
    name="Glyceraldehyde 3-phosphate",
    compartment='c')
DPG13=Metabolite(
    "DPG13",
    formula="",
    name="1,3-Diphosphoglycerate",
    compartment='c')

PG3=Metabolite(
    "PG3",
    formula="",
    name="3-Phosphoglycerate",
    compartment='c')

PG2=Metabolite(
    "PG2",
    formula="",
    name="2-Phosphoglycerate",
    compartment='c')

PEP=Metabolite(
    "PEP",
    formula="",
    name="Phosphoenolpyruvate",
    compartment='c')

PYR=Metabolite(
    "PYR",
    formula="",
    name="Pyruvate",
    compartment='c')

LAC=Metabolite(
    "LAC",
    formula="",
    name="Lactate",
    compartment='c')

NAD=Metabolite(
    "NAD",
    formula="",
    name="Nicotinamide adenine dinucleotide (oxidized)",
    compartment='c') 

NADH=Metabolite(
    "NADH",
    formula="",
    name="Nicotinamide adenine dinucleotide (reduced)",
    compartment='c')  

AMP=Metabolite(
    "AMP",
    formula="",
    name="Adenosine mono-phosphate",
    compartment='c')  

ADP=Metabolite(
    "ADP",
    formula="C10H15N5O10P2",
    name="Adenosine di-phosphate",
    compartment='c')

ATP=Metabolite(
    "ATP",
    formula="C10H12N5O13P3",
    name="Adenosine tri-phosphate",
    compartment='c')

PI=Metabolite(
    "PI",
    formula="",
    name="Inorganic phosphate",
    compartment='c')

H=Metabolite(
    "H",
    formula="H",
    name="Proton",
    compartment='c')

H2O=Metabolite(
    "H2O",
    formula="",
    name="Water",
    compartment='c')
```

```python
HK.add_metabolites({Glu:-1,ATP:-1,ADP:1,G6P:1,H:1})
PGI.add_metabolites({G6P:-1,F6P:1})
PFK.add_metabolites({F6P: , ATP: ,FDP: , ADP: ,H:})
TPI.add_metabolites({DHAP: , GAP:})
ALD.add_metabolites({FDP: , DHAP: , GAP:})
GAPDH.add_metabolites({GAP: ,NAD: , Pi:, DPG13: , NADH :, H:})
PGK.add_metabolites({DPG13:,  ADP: , PG3: , ATP:})
PGLM.add_metabolites({PG3:,  PG2:})
ENO.add_metabolites({PG2: ,PEP: , H2O:})
PK.add_metabolites({PEP: ,ADP: , H: ,PYR: , ATP:})
LDH.add_metabolites({PYR:,  NADH:,  H:,  LAC:, NAD:})
AMP.add_metabolites({AMP:})
APK.add_metabolites({ADP:, AMP: , ATP:})
PYR.add_metabolites({PYR :})
LAC.add_metabolites({LAC:})
ATP.add_metabolites({ATP: , H:2 , O: , ADP: ,Pi: , H:})
NADH.add_metabolites({NADH: ,NAD: , H:})
Gluin.add_metabolites({Glu:1})
AMPin.add_metabolites({AMP:})
H.add_metabolites({H:})
H2O.add_metabolites({H:2,O:})


```

```python
model.add_reactions({Gluin,HK,PGI,NADH,ATP,LAC,PYR,APK,AMPin,AMP,LDH,PK,ENO,H,H2O,PGLM,PGK,GAPDH,ALD,TPI,PFK})
```
